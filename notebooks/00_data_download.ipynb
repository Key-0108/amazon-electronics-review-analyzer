```python
import pandas as pd
import json

filepath = r"C:\Users\keert\Projects\Electronics.jsonl\Electronics.jsonl"
SAMPLE_EVERY = 400   # ~22M rows ÷ 400 = ~55K rows sampled evenly

records = []
with open(filepath, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i % SAMPLE_EVERY == 0:
            line = line.strip()
            if line:
                records.append(json.loads(line))

df = pd.DataFrame(records)
print(f"Loaded: {df.shape}")
print(df.columns.tolist())
print(df.head(2))
```
